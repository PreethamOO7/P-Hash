```python
def get_url(image_asset):

   image_asset = eval(image_asset)

   return image_asset["original"]["uri"]

def get_image(url):

   response = requests.get(url)

   im = Image.open(BytesIO(response.content)).convert("RGB")

   im.thumbnail(size = [500,500])

   #display(im)

   return im
```

```python
for i in range(13):
    ima= get_image(f1["IMAGE_URL"].loc[i])
    if(i%100==0):
        #data.to_csv("Ajio_Mens_Data_newpath.csv",index=False)
        print(i)
        
    jio_code = str(f1["P_JIOOCODE   "].loc[i])
    f = "F:\\Common_Brand_Ajio\\"+jio_code + ".jpg"
    ima.save(f)
    f1["new_path"].loc[i] = f
```
