```python
import numpy as np
import os,sys
from PIL import Image
import six
import imagehash
import pandas as pd
import operator
```

```python
df_ajio=pd.read_csv('E:/Temp/Common_Ajio.csv')
df_ajio.drop_duplicates(subset = ["PDP Url"], inplace = True)
df_ajio
```

```python
df_udaan = pd.read_csv('E:/compare/Common_Udaan1.csv')
df_udaan
```

```python
Udaan_dict = {}
ix = df_udaan.index
for i in ix:
    complete_path = df_udaan["new_path"].loc[i]
    url = df_udaan["Website_Url"].loc[i]
    Udaan_dict[url] = imagehash.phash(Image.open(complete_path))
```

```python
Ajio_dict = {}
ix = df_ajio.index
for i in ix:
    complete_path = df_ajio["new_path"].loc[i]
    url = df_ajio["PDP Url"].loc[i]
    Ajio_dict[url] = imagehash.phash(Image.open(complete_path))
```

```python
udaan_keys = Udaan_dict.keys()
ajio_keys = Ajio_dict.keys()
#l = [[0 for x in range(w)] for y in range(h)]  
l=[]
for h,i in enumerate(udaan_keys):
    print(i)
    k={}
    p1 = Udaan_dict[i]
    #print(p1)
    for j in ajio_keys:
        p2 = Ajio_dict[j]
        #print(j)
        val=p1-p2
        k[j] = val
    l.append(k)
```

```python
def get_closest_five(x):
    sorted_x=sorted(x.items(), key=operator.itemgetter(1))[:5]
    return sorted_x
```
